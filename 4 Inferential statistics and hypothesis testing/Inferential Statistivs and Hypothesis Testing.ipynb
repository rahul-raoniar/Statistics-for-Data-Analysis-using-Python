```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# rolling a die example (100 times)
one_h = np.random.randint(1, 7, 100)

one_h
```

```python
fig, ax = plt.subplots(figsize = (8, 4))

sns.histplot(x = one_h, bins = 6, ax=ax)
```

```python
# rolling a die example (1 million times)
one_m = np.random.randint(1, 7, 1000000)

fig, ax = plt.subplots(figsize = (8, 4))

sns.histplot(x = one_m, bins = 6, ax=ax)
```

```python
# Population mean and std
print(np.mean(one_m))
print(np.std(one_m))
```

### Taking 4 samples and calculating mean

```python
one_h.reshape(25, 4)
```

```python
# Reshaping it to 25 rows and 4 columns (total 100 sample)
# Taking mean for each row columnwise

mean_4 = one_h.reshape(25, 4).mean(axis = 1)

fig, ax = plt.subplots(figsize = (8, 4))

sns.histplot(x = mean_4, ax=ax)
ax.set_xlim(1,6)
```

```python
# Reshaping it to 25 rows and 4 columns (for total 1 million sample)
# Taking mean for each row columnwise

mean_4 = one_m.reshape(250000, 4).mean(axis = 1)

fig, ax = plt.subplots(figsize = (8, 4))

sns.histplot(x = mean_4, ax=ax, bins = 6)
ax.set_xlim(1,6)
```

```python
# Mean and std of sample means

print(np.mean(mean_4))
print(np.std(mean_4))
```

### Taking 100 samples and calculating mean

```python
# Reshaping it to 10000 rows and 100 columns (for total 1 million samples)
# Taking mean for each row columnwise (for 100 row samples)

mean_100 = one_m.reshape(10000, 100).mean(axis = 1)

fig, ax = plt.subplots(figsize = (8, 4))

sns.histplot(x = mean_100, ax=ax, bins = 6)
ax.set_xlim(1,6)
```

```python
# Mean and std of sample means

print(np.mean(mean_100))
print(np.std(mean_100))
```

### Testing the central limit theorem with bimodal distribution

```python
a = np.random.normal(100, 5, 1000)
b = np.random.normal(80, 5, 1000)

bm_data = np.append(a, b)
np.random.shuffle(bm_data) # sguffle the data

fig, ax = plt.subplots(figsize = (8, 4))

sns.histplot(x = bm_data, ax=ax)
```

```python
### Reshaping to take 10 sample

bm_data_2 = bm_data.reshape(1000, 2).mean(axis = 1)
bm_data_5 = bm_data.reshape(400, 5).mean(axis = 1)

fig, ax = plt.subplots(figsize = (8, 4), nrows = 2, ncols =1)
sns.histplot(x = bm_data_2, ax=ax[0])
sns.histplot(x = bm_data_5, ax=ax[1])
ax[0].set_title("2 sample mean")
ax[1].set_title("5 sample mean")

print("original mean", np.mean(bm_data))
print("2 sample mean", np.mean(bm_data_2))
print("5 sample mean", np.mean(bm_data_5))
print("")
print("original std", np.std(bm_data))
print("2 sample std", np.std(bm_data_2))
print("5 sample std",  np.std(bm_data_5))
```
